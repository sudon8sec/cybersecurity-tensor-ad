```python
import boto3
import sagemaker
from pathlib import Path
from tqdm import tqdm
```

```python
sess = sagemaker.Session()
bucket = sess.default_bucket()
s3_client = boto3.client('s3')
```

```python
data_dir = Path('/home/sagemaker-user/cybersecurity-tensor-ad/data/raw/cicids2017')

csv_files = list(data_dir.glob('*.csv'))

for csv_file in tqdm(csv_files):
    s3_key = f'cybersecurity-tensor-ad/raw/{csv_file.name}'
    s3_client.upload_file(str(csv_file), bucket, s3_key)

print(f"Location: s3://{bucket}/cybersecurity-tensor-ad/raw/")
```
